```python
def prospech(body):
    if body >= 40:
        return "Prospěl"
    else:
        return "Neprospěl"
```

```python

df = pandas.DataFrame({"body": [90, 20, 65, 75, 95, 45, 30, 10]})
df["prospěch"] = df["body"].apply(prospech)
df
```
